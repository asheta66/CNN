```python
import matplotlib.pyplot as plt

# Reset to default settings
plt.rcParams.update(plt.rcParamsDefault)
```

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.callbacks import LearningRateScheduler
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define your data directory path
data_dir = '/content/drive/My Drive/data4'
target_size = (128, 128)  # Desired resolution

# Check if the data directory exists
if not os.path.exists(data_dir):
    raise FileNotFoundError(f"Directory '{data_dir}' does not exist.")

# Get class names (subfolders)
classes = [cls for cls in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, cls))]
if len(classes) != 2:
    raise ValueError("This code is designed for exactly 2 classes. Please adjust the data directory or code accordingly.")

# Display sample image one from each class before and after resizing
def display_sample_images(data_dir, target_size, classes):
    fig, axes = plt.subplots(2, 2, figsize=(8, 8))
    for i, cls in enumerate(classes):
        cls_dir = os.path.join(data_dir, cls)
        img_file = os.listdir(cls_dir)[0]
        img_path = os.path.join(cls_dir, img_file)

        # Original image
        img = Image.open(img_path)
        axes[i, 0].imshow(img)
        axes[i, 0].set_title(f'{cls} - Original')
        axes[i, 0].axis('off')

        # Resized image
        resized_img = img.resize(target_size)
        axes[i, 1].imshow(resized_img)
        axes[i, 1].set_title(f'{cls} - Resized')
        axes[i, 1].axis('off')

    plt.tight_layout()
    plt.show()

display_sample_images(data_dir, target_size, classes)

# Prepare data
X, y = [], []
label_map = {cls: idx for idx, cls in enumerate(classes)}

for cls in classes:
    cls_dir = os.path.join(data_dir, cls)
    for img_file in os.listdir(cls_dir):
        img_path = os.path.join(cls_dir, img_file)
        img = load_img(img_path, target_size=target_size)
        img_array = img_to_array(img)
        X.append(img_array)
        y.append(label_map[cls])

X = np.array(X).astype('float32') / 255.0
y = np.array(y).astype('int')

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=22)

# Build and compile the CNN model
def build_compile_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'),
        MaxPooling2D((2, 2)),
        BatchNormalization(),

        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        BatchNormalization(),

        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        BatchNormalization(),

        Flatten(),
        Dense(128, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Model training with data augmentation
def scheduler(epoch, lr):
    if epoch < 10:
        return float(lr)
    else:
        return float(lr * tf.math.exp(-0.1))

lr_scheduler = LearningRateScheduler(scheduler)

datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

model = build_compile_model(input_shape=(target_size[0], target_size[1], 3), num_classes=len(classes))
history = model.fit(datagen.flow(X_train, y_train, batch_size=32),
                    epochs=30,
                    validation_data=(X_test, y_test),
                    callbacks=[lr_scheduler])

# Plot convergence curves
def plot_convergence_curves(history, filename='convergence_curves.png'):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
    ax1.plot(history.history['accuracy'], label='Train Accuracy')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Accuracy Convergence')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Accuracy')
    ax1.legend()
    ax1.grid()

    ax2.plot(history.history['loss'], label='Train Loss')
    ax2.plot(history.history['val_loss'], label='Validation Loss')
    ax2.set_title('Loss Convergence')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Loss')
    ax2.legend()
    ax2.grid()

    plt.tight_layout()
    plt.savefig(filename, bbox_inches='tight')
    plt.show()
    plt.close()

plot_convergence_curves(history)

# Predict and evaluate
train_predictions = np.argmax(model.predict(X_train), axis=1)
test_predictions = np.argmax(model.predict(X_test), axis=1)

# Compute metrics
metrics = []

def compute_metrics(true_labels, predictions):
    return {
        'Accuracy': accuracy_score(true_labels, predictions),
        'Precision': precision_score(true_labels, predictions, average='weighted'),
        'Recall': recall_score(true_labels, predictions, average='weighted'),
        'F1 Score': f1_score(true_labels, predictions, average='weighted')
    }

metrics.append({'Dataset': 'Training', **compute_metrics(y_train, train_predictions)})
metrics.append({'Dataset': 'Testing', **compute_metrics(y_test, test_predictions)})

metrics_df = pd.DataFrame(metrics)
print(metrics_df)

# Save metrics to Excel
metrics_df.to_excel('performance_metrics.xlsx', index=False, sheet_name='Performance Metrics')

# Plot confusion matrices
def plot_confusion_matrices(cm_train, cm_test, class_names, filename='confusion_matrices.png'):
    fig, axes = plt.subplots(1, 2, figsize=(8, 3))
    sns.heatmap(cm_train, annot=True, fmt='d', ax=axes[0], cmap='Blues', xticklabels=class_names, yticklabels=class_names, cbar=False)
    axes[0].set_title('Training Confusion Matrix')
    axes[0].set_xlabel('Predicted labels')
    axes[0].set_ylabel('True labels')

    sns.heatmap(cm_test, annot=True, fmt='d', ax=axes[1], cmap='Blues', xticklabels=class_names, yticklabels=class_names, cbar=False)
    axes[1].set_title('Testing Confusion Matrix')
    axes[1].set_xlabel('Predicted labels')
    axes[1].set_ylabel('True labels')

    plt.tight_layout()
    plt.savefig(filename, bbox_inches='tight')
    plt.show()
    plt.close()

cm_train = confusion_matrix(y_train, train_predictions)
cm_test = confusion_matrix(y_test, test_predictions)

plot_confusion_matrices(cm_train, cm_test, classes)


# Build and compile the CNN model
def build_compile_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', input_shape=input_shape, padding='same'),
        MaxPooling2D((2, 2)),
        BatchNormalization(),

        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        BatchNormalization(),

        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        BatchNormalization(),

        Conv2D(256, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        BatchNormalization(),

        Flatten(),
        Dense(512, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Prepare data
X, y = [], []
label_map = {cls: idx for idx, cls in enumerate(classes)}

for cls in classes:
    cls_dir = os.path.join(data_dir, cls)
    for img_file in os.listdir(cls_dir):
        img_path = os.path.join(cls_dir, img_file)
        img = load_img(img_path, target_size=target_size)
        img_array = img_to_array(img)
        X.append(img_array)
        y.append(label_map[cls])

X = np.array(X).astype('float32') / 255.0
y = np.array(y).astype('int')

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=22)

# Build and compile model
input_shape = (target_size[0], target_size[1], 3)
num_classes = len(classes)
model = build_compile_model(input_shape, num_classes)

# Learning Rate Scheduler function
def scheduler(epoch, lr):
    if epoch < 10:
        return float(lr)
    else:
        return float(lr * tf.math.exp(-0.1))

lr_scheduler = LearningRateScheduler(scheduler)

# Data augmentation
datagen = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Train the model with data augmentation
history = model.fit(datagen.flow(X_train, y_train, batch_size=32),
                    epochs=30,
                    validation_data=(X_test, y_test),
                    callbacks=[lr_scheduler])

# Initialize a list to store metrics for saving
metrics = []

# Predict and evaluate
train_predictions = np.argmax(model.predict(X_train), axis=1)
test_predictions = np.argmax(model.predict(X_test), axis=1)

# Compute metrics for training data
train_accuracy = accuracy_score(y_train, train_predictions)
train_precision = precision_score(y_train, train_predictions, average='weighted')
train_recall = recall_score(y_train, train_predictions, average='weighted')
train_f1 = f1_score(y_train, train_predictions, average='weighted')

# Store training metrics in the list
metrics.append({
    'Dataset': 'Training',
    'Accuracy': train_accuracy,
    'Precision': train_precision,
    'Recall': train_recall,
    'F1 Score': train_f1
})

# Compute metrics for testing data
test_accuracy = accuracy_score(y_test, test_predictions)
test_precision = precision_score(y_test, test_predictions, average='weighted')
test_recall = recall_score(y_test, test_predictions, average='weighted')
test_f1 = f1_score(y_test, test_predictions, average='weighted')

# Store testing metrics in the list
metrics.append({
    'Dataset': 'Testing',
    'Accuracy': test_accuracy,
    'Precision': test_precision,
    'Recall': test_recall,
    'F1 Score': test_f1
})

# Create a DataFrame from the metrics list
metrics_df = pd.DataFrame(metrics)

# Display metrics DataFrame
print(metrics_df)

# Define filename for Excel
excel_filename = 'performance_metrics.xlsx'

# Save the DataFrame to an Excel file
metrics_df.to_excel(excel_filename, index=False, sheet_name='Performance Metrics')

print(f"Metrics saved to {excel_filename}")

# Alternative to plot_model: Print model summary
with open('model_summary.txt', 'w') as f:
    with tf.device('/CPU:0'):
        model.summary(print_fn=lambda x: f.write(x + '\n'))

print(f"Model summary saved to model_summary.txt")

# Plot confusion matrices
def plot_confusion_matrices(cm_train, cm_test, class_names, filename='confusion_matrices.png'):
    fig, axes = plt.subplots(1, 2, figsize=(6, 4))  # Reduced size
    sns.heatmap(cm_train, annot=True, fmt='d', ax=axes[0], cmap='Blues', xticklabels=class_names, yticklabels=class_names, cbar=False, annot_kws={"size": 9})
    axes[0].set_title('Training Confusion Matrix', fontsize=10)
    axes[0].set_xlabel('Predicted labels', fontsize=9)
    axes[0].set_ylabel('True labels', fontsize=9)

    sns.heatmap(cm_test, annot=True, fmt='d', ax=axes[1], cmap='Blues', xticklabels=class_names, yticklabels=class_names, cbar=False, annot_kws={"size": 9})
    axes[1].set_title('Testing Confusion Matrix', fontsize=10)
    axes[1].set_xlabel('Predicted labels', fontsize=9)
    axes[1].set_ylabel('True labels', fontsize=9)

    plt.tight_layout()
    plt.savefig(filename, bbox_inches='tight')
    plt.show()
    plt.close()

# Compute confusion matrices
cm_train = confusion_matrix(y_train, train_predictions)
cm_test = confusion_matrix(y_test, test_predictions)

# Plot and save confusion matrices
plot_confusion_matrices(cm_train, cm_test, classes, filename=f'confusion_matrices_{target_size[0]}x{target_size[1]}.png')

# Plot convergence curves
def plot_convergence_curves(history, filename='convergence_curves.png'):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 4))  # Reduced size
    ax1.plot(history.history['accuracy'], label='Train Accuracy')
    ax1.plot(history.history['val_accuracy'], label='Validation Accuracy')
    ax1.set_title('Accuracy Convergence', fontsize=10)
    ax1.set_xlabel('Epoch', fontsize=9)
    ax1.set_ylabel('Accuracy', fontsize=9)
    ax1.legend()
    ax1.grid()

    ax2.plot(history.history['loss'], label='Train Loss')
    ax2.plot(history.history['val_loss'], label='Validation Loss')
    ax2.set_title('Loss Convergence', fontsize=10)
    ax2.set_xlabel('Epoch', fontsize=9)
    ax2.set_ylabel('Loss', fontsize=9)
    ax2.legend()
    ax2.grid()

    plt.tight_layout()
    plt.savefig(filename, bbox_inches='tight')
    plt.show()
    plt.close()

# Plot and save convergence curves
plot_convergence_curves(history, filename=f'convergence_curves_{target_size[0]}x{target_size[1]}.png')
```
